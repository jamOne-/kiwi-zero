```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras import layers
```

```python
def get_model(input_shape):
    model = tf.keras.Sequential()
    model.add(layers.Dense(1, activation="sigmoid", use_bias=False, input_shape=input_shape, name="output"))

    return model
```

```python
model = get_model((128,))
print(model.count_params())
print(model.get_layer('output').get_weights())

# while True:
#     weights = np.array(map(float, input().split(" ")))
#     Xs_length = int(input())
#     Xs = [list(map(float, input().split(" "))) for i in range(Xs_length)]
#     ys = [list(map(float, input().split(" "))) for i in range(Xs_length)]

#     Xs = np.array(Xs)
#     ys = np.array(ys)
```
